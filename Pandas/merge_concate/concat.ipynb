```python
import pandas as pd # type: ignore

india_whether = pd.DataFrame({
  "city": ["mumbai", "delhi", "banglore"],
    "temperature": [32, 45, 30],
    "humidity": [80, 60, 78]
})

india_whether
```

```python
us_whether = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
        "temperature": [21, 14, 35],
        "humidity": [68, 65, 75]
    })

us_whether
```

```python
df_concat = pd.concat([india_whether, us_whether], ignore_index=True)
df_concat
```

```python
df_concat = pd.concat([india_whether, us_whether], keys=["india", "us"])
df_concat
```

```python
df_concat.loc["india"] 
df_concat.loc["us"]
```

```python
df_temp = pd.DataFrame({
    "city": ["mumbai", "delhi", "banglore"],
    "temperature": [32, 45, 30]
},index=[0,1,2])
df_temp
```

```python
df_windspeed = pd.DataFrame({   
    "city": ["mumbai", "delhi", "banglore"],
    "windspeed": [7, 12, 9]
},index=[0,1,2])
df_windspeed
```

```python
df_new_concate = pd.concat([df_temp, df_windspeed], axis=1)
df_new_concate
```

```python
df1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando", "baltimore"],
    "temperature": [21, 14, 35, 38],
})
```

```python
df2 = pd.DataFrame({
    "city": ["chicago", "new york", "orlando", "san francisco"],
    "humidity": [65, 68, 75, 71],
})
```

```python
df_merge = pd.merge(df1, df2, on="city")
df_merge
```

```python
df_merge = pd.merge(df1, df2, on="city",how="outer")
df_merge
```

```python
df_merge = pd.merge(df1, df2, on="city",how="left")
df_merge
```

```python
df_merge = pd.merge(df1, df2, on="city",how="inner")
df_merge
```
